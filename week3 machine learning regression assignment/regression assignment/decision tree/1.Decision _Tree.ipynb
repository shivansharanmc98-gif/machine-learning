```python
#importing the pandas Libraies
import pandas as pd
```

```python
# Reading the Dataset
dataset=pd.read_csv("insurance_pre.csv")
```

```python
#output of the dataset
dataset
```

```python
#data preprocessing using get_dummies to change catagorical data to numerical data which can understande by model
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
#independend data x values
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
```

```python
independent
```

```python
#dependent data y values
dependent=dataset[["charges"]]
```

```python
dependent
```

```python
#split into training set and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30,random_state=0)
```

```python
#standardizing the input data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeRegressor.html
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(criterion='poisson', splitter='best', max_features = "log2")
regressor=regressor.fit(X_train,y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
#loading and saving the model using pickel
import pickle
filename="finalized_model_decisiontree_linear.sav"
pickle.dump(regressor,open(filename,'wb'))
```
