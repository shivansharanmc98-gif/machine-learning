```python
#importing the Libraies
import pandas as pd
```

```python
# Reading the Dataset
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
```

```python
independent
```

```python
dependent=dataset[["charges"]]
```

```python
dependent
```

```python
#split into training set and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30,random_state=0)
```

```python
#data preprocessing standardizing the input data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeRegressor.html
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor

regressor = XGBRegressor(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=4
)
```

```python
# Train
regressor.fit(X_train, y_train)
```

```python
#prediction
y_pred=regressor.predict(X_test)
```

```python
#loading y_test,y_pred in r2 matrics value
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
#loading and saving the model using pickel
import pickle
filename="finalized_model_xg_linear.sav"
pickle.dump(regressor,open(filename,'wb'))
```

```python
loaded_model=pickle.load(open("finalized_model_xg_linear.sav",'rb'))
result=loaded_model.predict([[25,20,2,1,0]])
```
