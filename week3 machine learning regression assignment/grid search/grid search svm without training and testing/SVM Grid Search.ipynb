```python
#importing the pandas Libraies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Reading the Dataset
dataset = pd.read_csv('insurance_pre.csv')
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
indep=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
dep=dataset['charges']
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
param_grid = {'kernel':['rbf','poly','sigmoid','linear'],'C':[10,100,1000,2000],'gamma': ["scale", 0.001, 0.0001]}
 # fitting the model for grid search
grid = GridSearchCV(SVR(), param_grid, refit = True, verbose = 3,n_jobs=-1,cv=6)
grid.fit(indep, dep)
re=grid.cv_results_
       
```

```python
table=pd.DataFrame.from_dict(re)
```

```python
table
```

```python
age_input=float(input("Age:"))
bmi_input=float(input("BMI:"))
children_input=float(input("Children:"))
sex_male_input=int(input("Sex Male 0 or 1:"))
smoker_yes_input=int(input("Smoker Yes 0 or 1:"))
```

```python
# Same order as X columns
Future_Prediction = grid.predict([[age_input, bmi_input, children_input,
                                   sex_male_input, smoker_yes_input]])

print("Future_Prediction = {}".format(Future_Prediction))
```
