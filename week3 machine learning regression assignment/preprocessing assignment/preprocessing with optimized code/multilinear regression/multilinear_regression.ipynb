```python
#importing the pandas Libraies
import pandas as pd
```

```python
# Reading the Dataset
dataset=pd.read_csv("insurance_pre.csv")
```

```python
#output of the dataset
dataset
```

```python
#data preprocessing using get_dummies to change catagorical data to numerical data which can understande by model
dataset = pd.get_dummies(dataset, drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
#independend data x values
independent=dataset[['age', 'bmi', 'children',  'sex_male', 'smoker_yes']]
```

```python
#dependent data y values
dependent=dataset[["charges"]]
```

```python
#split into training set and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30,random_state=0)
```

```python
#standardizing the input data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#standardizing the output data
scy = StandardScaler()
y_train = scy.fit_transform(y_train)
y_test = scy.transform(y_test)
```

```python
# fitting the model for LinearRegression 
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)
```

```python
weight=regressor.coef_
weight
```

```python
bais=regressor.intercept_
bais
```

```python
#prediction by loading X_test vlaues to the predict
y_pred=regressor.predict(X_test)
```

```python
#loading y_test,y_pred in r2 matrics valu
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
#loading and saving the model using pickel
import pickle
```

```python
pickle.dump(regressor, open("finalized_model.sav", "wb"))
pickle.dump(sc, open("sc_X.pkl", "wb"))
pickle.dump(scy, open("sc_y.pkl", "wb"))
```
