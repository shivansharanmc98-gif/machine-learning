```python
#loading and saving the model using pickel
import pickle

# Load model & scalers
regressor= pickle.load(open("finalized_model.sav", "rb"))
sc = pickle.load(open("sc_X.pkl", "rb"))
scy = pickle.load(open("sc_y.pkl", "rb"))

# New input (example)
new_data = [[25, 20, 2, 1, 0]]  
# age, bmi, children, sex_male, smoker_yes

# Predict
scaled = sc.transform(new_data)
pred = regressor.predict(scaled)
final_output = scy.inverse_transform(pred)

print("Predicted Charges:", final_output[0][0])
```
